## Upload dataset lên

```python
rf = Roboflow(api_key="API_KEY")
project = rf.workspace("fruit-detector-blc0f").project("skin-cancer-detection-wfldq-qtzod")
version = project.version(2)
dataset = version.download("folder")
```

```python
from ultralytics import YOLO
from google.colab import runtime
import os

model = YOLO("yolo26x-cls.pt")

results = model.train(
    data=dataset.location,
    epochs=60,
    imgsz=800,
    batch=32,
    workers=6,
    amp=True,

    dropout=0.3,
    label_smoothing=0.1,
    weight_decay=0.001,
    optimizer='AdamW',
    lr0=0.0006,
    cos_lr=True,

    project=SAVE_DIR,
    name="train",
    plots=True,
    save_period=10,
)

print("Training hoàn tất!")
runtime.unassign()
```
